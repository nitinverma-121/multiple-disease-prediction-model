```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv('heart.csv')
data.head()
```

```python
data.groupby('target').mean()
```

```python
features= data.drop('target', axis= 1)
```

```python
features
```

```python
data.max()
```

```python
target= data['target']
target
```

```python
x_train, x_test, y_train, y_test= train_test_split(features, target, test_size= 0.33, stratify= target, random_state= 10)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(max_iter= 930)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
import pickle
pickle.dump(model, open('heart_model.pkl', 'wb'))
```
