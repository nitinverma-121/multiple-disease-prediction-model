```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as ptr 
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv('parkinsons.csv')
data.head()
```

```python
features= data.drop(columns=['name', 'status'])
features
```

```python
data.max()
```

```python
data.min()
```

```python
target= data['status']
target
```

```python
scaler= StandardScaler()
scaler.fit(features)
import pickle
pickle.dump(scaler, open('par_scaler.pkl', 'wb'))
```

```python
features= scaler.transform(features)
```

```python
features
```

```python
x_train, x_test, y_train, y_test= train_test_split(features, target, test_size= 0.33, random_state= 10)
```

```python
#svm
from sklearn.svm import SVC

model= SVC(C= 7.5)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
import pickle
pickle.dump(model, open('par_model.pkl', 'wb'))
```
